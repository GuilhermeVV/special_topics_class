### Pillow (PIL)
<p>The Python Imaging Library, or PIL for short, is one of the core libraries for image manipulation in Python. Unfortunately, its development has stagnated, with its last release in 2009.</p>
<p>However, there's an actively-developed fork of PIL called Pillow (https://pillow.readthedocs.io/).</p>

### Imports

````python
from PIL import Image

import numpy as np
import matplotlib.pyplot as plt
````

### Reading and inspecting the image

<img src="../../resources/img_rep.png"/>

````python
img_path = '../../sample_images/machine-learning-meetup.png'
````

````python
img = Image.open(img_path)
````

### Verifying shape (width, height, channels)
Note that when using Pillow's open method, we actually do not get a numpy array.

````python
print(img.shape)
````

We actually have a Pillow object

````python
print(img)
````

### Converting to numpy array

````python
img_array = np.asarray(img)
````

### Verifying shape (width, height, channels)
We can now verify the shape, but notice again we're getting a 4th channel (RGBA).

````python
print(img_array.shape)
````

But the type is uint8

````python
print(img_array.dtype)
````

### Showing the image
It works with the original `img`

````python
plt.imshow(img)
````

As well as with `img_array`

````python
plt.imshow(img_array)
````

### Converting to RGB

````python
img_rgb = img.convert(mode='RGB')
````

Notice that the background is now black when we convert to RGB.

````python
plt.imshow(img_rgb)
````

````python
img_rgb_array = np.array(img_rgb)
````

````python
print(img_rgb_array.shape)
````

### Removing the black background

There are multiple ways to solve this, but one way is shown below.

````python
background = Image.new('RGBA', img.size, (255,255,255))
img_rgb2 = Image.alpha_composite(background, img)

img_rgb2_array = np.asarray(img_rgb2)
````

````python
plt.imshow(img_rgb2_array)
````

### Resizing the image

````python
img_rgb2_resized = img_rgb2.resize((300, 300))
img_rgb2_resized_array = np.array(img_rgb2_resized)
````

````python
plt.imshow(img_rgb2_resized_array)
````

````python
print(img_rgb2_resized_array.shape)
````

### Converting to grayscale

````python
img_gray = img.convert(mode='L')
````

````python
plt.imshow(img_gray, cmap='gray')
````
